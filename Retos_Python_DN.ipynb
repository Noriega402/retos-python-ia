## Funcion que tome dos numeros como argumentos y devuelva el numero mayor

```python
def numero_mayor(x,y):
  if x > y:
    return x
  elif x == y:
    print('Ambos son el mismo numero')
    return 0
  else:
    return y

print(numero_mayor(3,1))
```
